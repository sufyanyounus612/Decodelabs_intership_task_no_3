```python
# =========================================================
# DATA ANALYTICS PROJECT 3
# SQL DATA ANALYSIS USING PYTHON + SQLITE
# =========================================================

# =========================================================
# IMPORT REQUIRED LIBRARIES
# =========================================================

import pandas as pd
import sqlite3

# =========================================================
# LOAD DATASET
# =========================================================

try:
    # Dataset File Path
    FILE_PATH = "Dataset for Data Analytics - Sheet1.csv"

    # Read CSV File
    df = pd.read_csv(FILE_PATH)

    print("=" * 70)
    print(" DATASET LOADED SUCCESSFULLY ")
    print("=" * 70)

except FileNotFoundError:
    print(" Dataset file not found.")
    exit()

# =========================================================
# DISPLAY BASIC DATASET INFORMATION
# =========================================================

print("\n FIRST 5 ROWS")
print(df.head())

print("\n DATASET INFORMATION")
print(df.info())

print("\n DATASET SHAPE")
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

print("\n COLUMN NAMES")
print(df.columns.tolist())

# =========================================================
# CREATE SQLITE CONNECTION
# =========================================================

connection = sqlite3.connect(":memory:")

# =========================================================
# STORE DATAFRAME INTO SQL TABLE
# =========================================================

TABLE_NAME = "orders"

df.to_sql(TABLE_NAME, connection, if_exists="replace", index=False)

print("\n SQL Table Created Successfully")

# =========================================================
# FUNCTION TO EXECUTE SQL QUERIES
# =========================================================

def execute_query(title, query):

    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

    try:
        result = pd.read_sql_query(query, connection)

        if result.empty:
            print(" No Records Found")
        else:
            print(result)

    except Exception as error:
        print(f" Query Error: {error}")

# =========================================================
# QUERY 1 — FIRST 10 RECORDS
# =========================================================

query_1 = """
SELECT *
FROM orders
LIMIT 10;
"""

execute_query(
    " QUERY 1 — FIRST 10 RECORDS",
    query_1
)

# =========================================================
# QUERY 2 — DELIVERED ORDERS
# =========================================================

query_2 = """
SELECT OrderID,
       Product,
       Quantity,
       TotalPrice
FROM orders
WHERE OrderStatus = 'Delivered';
"""

execute_query(
    " QUERY 2 — DELIVERED ORDERS",
    query_2
)

# =========================================================
# QUERY 3 — TOP EXPENSIVE ORDERS
# =========================================================

query_3 = """
SELECT OrderID,
       Product,
       TotalPrice
FROM orders
ORDER BY TotalPrice DESC
LIMIT 10;
"""

execute_query(
    " QUERY 3 — TOP 10 MOST EXPENSIVE ORDERS",
    query_3
)

# =========================================================
# QUERY 4 — TOTAL ORDERS BY PRODUCT
# =========================================================

query_4 = """
SELECT Product,
       COUNT(*) AS Total_Orders
FROM orders
GROUP BY Product
ORDER BY Total_Orders DESC;
"""

execute_query(
    " QUERY 4 — TOTAL ORDERS BY PRODUCT",
    query_4
)

# =========================================================
# QUERY 5 — TOTAL SALES BY PRODUCT
# =========================================================

query_5 = """
SELECT Product,
       ROUND(SUM(TotalPrice), 2) AS Total_Revenue
FROM orders
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""

execute_query(
    " QUERY 5 — TOTAL REVENUE BY PRODUCT",
    query_5
)

# =========================================================
# QUERY 6 — AVERAGE PRICE BY PRODUCT
# =========================================================

query_6 = """
SELECT Product,
       ROUND(AVG(UnitPrice), 2) AS Average_Price
FROM orders
GROUP BY Product
ORDER BY Average_Price DESC;
"""

execute_query(
    " QUERY 6 — AVERAGE PRODUCT PRICE",
    query_6
)

# =========================================================
# QUERY 7 — PAYMENT METHOD ANALYSIS
# =========================================================

query_7 = """
SELECT PaymentMethod,
       COUNT(*) AS Total_Transactions,
       ROUND(SUM(TotalPrice), 2) AS Revenue
FROM orders
GROUP BY PaymentMethod
ORDER BY Revenue DESC;
"""

execute_query(
    " QUERY 7 — PAYMENT METHOD ANALYSIS",
    query_7
)

# =========================================================
# QUERY 8 — ORDER STATUS REPORT
# =========================================================

query_8 = """
SELECT OrderStatus,
       COUNT(*) AS Total_Orders
FROM orders
GROUP BY OrderStatus
ORDER BY Total_Orders DESC;
"""

execute_query(
    " QUERY 8 — ORDER STATUS REPORT",
    query_8
)

# =========================================================
# QUERY 9 — REFERRAL SOURCE ANALYSIS
# =========================================================

query_9 = """
SELECT ReferralSource,
       COUNT(*) AS Customers,
       ROUND(SUM(TotalPrice), 2) AS Revenue
FROM orders
GROUP BY ReferralSource
ORDER BY Revenue DESC;
"""

execute_query(
    " QUERY 9 — REFERRAL SOURCE PERFORMANCE",
    query_9
)

# =========================================================
# QUERY 10 — HAVING CLAUSE
# =========================================================

query_10 = """
SELECT Product,
       ROUND(SUM(TotalPrice), 2) AS Total_Sales
FROM orders
GROUP BY Product
HAVING Total_Sales > 100000
ORDER BY Total_Sales DESC;
"""

execute_query(
    " QUERY 10 — PRODUCTS WITH SALES > 100000",
    query_10
)

# =========================================================
# FINAL INSIGHTS
# =========================================================

print("\n" + "=" * 70)
print(" FINAL BUSINESS INSIGHTS ")
print("=" * 70)

print("""
1. SQL queries simplify business data analysis.
2. GROUP BY helps analyze category-wise performance.
3. SUM() identifies revenue-generating products.
4. AVG() calculates average pricing insights.
5. ORDER BY highlights top-performing records.
6. HAVING filters grouped business results.
""")

# =========================================================
# CLOSE CONNECTION
# =========================================================

connection.close()

print("\n SQL Connection Closed")
print(" PROJECT COMPLETED SUCCESSFULLY")
print("=" * 70)
```
